```python
# Je créée un dataframe par indicateur dans l'objectif de les regrouper dans un seul DF, et accorder un score aux pays qui 
# ont des bon niveaux sur plusieurs indicateurs. 

top10_dfs = {}  # je créee un dictionnaire pour stocker les différents DFs.

indicators = ['Tertiary enrol %', 'UpperSec enrol %', 'GrossGrad 6&7ter', 'Internet', 'Ter5 enrol']

for i in indicators: 
    TOP10 = DATA[i].sort_values(ascending=False).head(10)
    df_top10 = TOP10.reset_index()
    df_top10.columns = ['Country Name', i]
    top10_dfs[i] = df_top10  # Je stocke les DFs dans le dictionnaire

display(top10_dfs['Internet']) # Je teste pour voir si ma boucle fonctionne 
display(top10_dfs['Tertiary enrol %']) # Je teste pour voir si ma boucle fonctionne 

# Je créee mes DFs : 

DF_net = top10_dfs['Internet']
DF_net

DF_ter = top10_dfs['Tertiary enrol %']
DF_ter

DF_upsec = top10_dfs['UpperSec enrol %']
DF_upsec

DF_GG67 = top10_dfs['GrossGrad 6&7ter']
DF_GG67

DF_ter5 = top10_dfs['Ter5 enrol']
DF_ter5
```

```python
indicators = ['Tertiary enrol %', 'UpperSec enrol %', 'GrossGrad 6&7ter', 'Internet', 'Ter5 enrol']

DF4['SCORE'] = DF4[indicators].sum(axis=1)
DF4.sort_values('SCORE', ascending = False)

# Pour le score j'opte simplement pour la somme de tous les indicateurs. Certes ce score en l'état n'a pas d'explication statistique
# mais il permet de classer les différents pays. Je vais faire également un classment par Moyenne, mais le résultat final sera similaire.
```

```python
DF4['Moyenne'] = (DF4['SCORE']/5)
TOP_global = DF4.sort_values('Moyenne', ascending = False)
TOP_global

# classement par moyenne des indicateurs pour les pays.
```
